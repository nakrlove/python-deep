```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import folium
```

```python
seoul_map=folium.Map(location=[37.55,126.98],zoom_start=12)
df = pd.read_excel("서울지역대학교위치.xlsx")
for name,lat,lng in zip(df['대학교'],df['위도'], df['경도']):
    folium.Marker([lat,lng],tooltip=name).add_to(seoul_map)
    
```

```python
seoul_map=folium.Map(location=[37.55,126.98],zoom_start=12)

for name,lat,lng in zip(df['대학교'],df['위도'], df['경도']):
    # folium.Marker([lat,lng],tooltip=name).add_to(seoul_map)
    folium.CircleMarker([lat,lng],color='white',fill=True,fill_color='red',fill_opacity=0.6).add_to(seoul_map)
```
